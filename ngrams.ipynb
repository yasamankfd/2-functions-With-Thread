```python
import nltk
from nltk import ngrams
from collections import Counter
import random
import string  # Import string module for generating random email addresses
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Download the punkt tokenizer data
nltk.download('punkt')

def compute_ngrams(corpus, n):
    tokens = nltk.word_tokenize(corpus)
    ngrams_list = list(ngrams(tokens, n))
    return ngrams_list

def compare_corpora_stats(corpus1, corpus2):
    n = 2  # Bigrams
    ngrams_list1 = compute_ngrams(corpus1, n)
    ngrams_list2 = compute_ngrams(corpus2, n)

    # Compute statistics for each corpus
    most_common_ngrams1 = [ngram for ngram, freq in Counter(ngrams_list1).most_common(10)]
    most_common_ngrams2 = [ngram for ngram, freq in Counter(ngrams_list2).most_common(10)]

    return most_common_ngrams1, most_common_ngrams2, ngrams_list1, ngrams_list2

def generate_random_sentence(ngrams_list):
    random_sentence = []


    while True:
        # Select a random starting word from the ngrams_list
        random_start = random.choice([ngram for ngram in ngrams_list if ngram[0][0].isupper()])
        random_sentence.extend(random_start)


        # Select the next word based on the previous word in the sentence
        while True:
            # Find all ngrams that start with the last word in the random_sentence
            candidate_ngrams = [ngram for ngram in ngrams_list if ngram[0] == random_sentence[-1]]
            # If no candidate ngrams are found, end the sentence
            if not candidate_ngrams:
                return ' '.join(random_sentence)
            # Select a random next word from the candidate ngrams
            next_word = random.choice(candidate_ngrams)
            random_sentence.append(next_word[1])
            # Check if the sentence ends with a period
            if next_word[1].endswith('.'):
                return ' '.join(random_sentence)

def generate_email_like_text():
    msg = MIMEMultipart()
    msg['From'] = generate_email_address()
    msg['To'] = generate_email_address()
    msg['Subject'] = 'Generated Email'
    body = generate_random_email_body()
    msg.attach(MIMEText(body, 'plain'))
    return msg.as_string()

def generate_email_address():
    username = ''.join(random.choices(string.ascii_letters, k=8))  # Generate random username
    domain = ''.join(random.choices(string.ascii_lowercase, k=5)) + '.com'  # Generate random domain
    return f"{username}@{domain}"

def generate_random_email_body():
    body = 'This is a randomly generated email-like text.\n'
    num_sentences = random.randint(3, 7)  # Random number of sentences in the email body
    for _ in range(num_sentences):
        sentence_length = random.randint(10, 30)  # Random length of each sentence
        sentence = ' '.join(random.choices(string.ascii_letters, k=sentence_length))
        body += sentence.capitalize() + '.\n'
    return body

# Generate email-like text for corpus 1 and corpus 2
email_text1 = generate_email_like_text()
email_text2 = generate_email_like_text()

# Save email-like text to files
with open('corpus1.txt', 'w') as file:
    file.write(email_text1)

with open('corpus2.txt', 'w') as file:
    file.write(email_text2)

# Read corpus text from files
def read_corpus_from_file(filename):
    with open(filename, 'r') as file:
        return file.read()

corpus1 = read_corpus_from_file('corpus1.txt')
corpus2 = read_corpus_from_file('corpus2.txt')

print("corpus 1:", corpus1)
print("corpus 2:", corpus2)

most_common_ngrams1, most_common_ngrams2, ngrams_list1, ngrams_list2 = compare_corpora_stats(corpus1, corpus2)
print("Most common ngrams in corpus 1:", most_common_ngrams1)
print("Most common ngrams in corpus 2:", most_common_ngrams2)

random_sentence1 = generate_random_sentence(ngrams_list1)
random_sentence2 = generate_random_sentence(ngrams_list2)
print("Random sentence from corpus 1:", random_sentence1)
print("Random sentence from corpus 2:", random_sentence2)
```
